```python
def GTP_APP(text):
    res=openai.Completion.create(
    model='text-davinci-003',
    prompt=text,
    max_tokens=500,
    temperature=0)
    return res.choices[0].text
```

```python
GTP_APP('what is NLP')[2:]
```

```python
GTP_APP('Please explain RNN model')[2:]
```

```python
def Create_Images(text):
    com = openai.Image.create(prompt=text, n=1, size="512x512")
    url = com['data'][0]['url']
    image_data = urllib.request.urlopen(url).read()
    image_file = io.BytesIO(image_data)
    image = Image.open(image_file)
    plt.imshow(image)
    plt.show()
```

```python
Create_Images('a football match')
```

```python
Create_Images('Build desktop setup with chair and green plant on right side of table')
```

<h2><center><strong>Fine tuning the model with some instructions for sentiment analysis</strong></center></h2>
        
</div>

```python
def training_data(data):
    return """Please predict text as positive or negative.
    text: You are a bad man.
    sentiment: negative
    text: You are good man.
    sentiment: positive
    text: {}
    sentiment:""".format(data.capitalize())

def Sentiment_APP(text):
    res=openai.Completion.create(
    model='text-davinci-003',
    prompt=training_data(text),
    max_tokens=20)
    return res.choices[0].text
```

```python
Sentiment_APP('You are a good man.')
```

```python
Sentiment_APP('You are damn wasting my time!')
```
